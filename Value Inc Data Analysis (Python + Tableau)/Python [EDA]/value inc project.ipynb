## <center>**Importing Necessary Libraries**

```python
import pandas as pd
```

## <center>**Reading the Dataset**

```python
data = pd.read_csv('transaction.csv', sep=';')
data.head()
```

## <center>**Data Definition**

```python
data.info()
```

## <center>**Summary of the dataset**

```python
data.describe()
```

## Replacing Year 2028 with 2020

```python
data['Year'] = data['Year'].replace(2028, 2020)
```

## <center>**Calculating new measures**

```python
data['CostPerTransactions'] = data['CostPerItem'] * data['NumberOfItemsPurchased']
data['SellingPricePerTransactions'] = data['SellingPricePerItem'] * data['NumberOfItemsPurchased']
data['ProfitPerItem'] = data['SellingPricePerItem'] - data['CostPerItem']
data['ProfitPerTransactions'] = data['ProfitPerItem'] * data['NumberOfItemsPurchased']
data['MarkupPerItem'] = ((data['SellingPricePerItem'] - data['CostPerItem']) / data['CostPerItem'])
data['MarkupPerTransactions'] = data['MarkupPerItem'] * data['NumberOfItemsPurchased']
```

```python
data.head()
```

```python
data = round(data, 2)
data.head()
```

## <center>**Working with dates**

```python
data.info()
```

```python
data['Day'] = data['Day'].astype(str)
data['Year'] = data['Year'].astype(str)
```

```python
data['Date'] = data['Day'] + '-' + data['Month'] + '-' + data['Year']
data.head()
```

## <center>**Some REGEX Operations (Split, replace, lower)**
### *Splitting the column 'ClientKeywords'*

```python
split_col = data['ClientKeywords'].str.split(',', expand = True)
```

```python
data['ClientAge'] = split_col[0]
data['ClientType'] = split_col[1]
data['ClientDuration'] = split_col[2]
```

```python
data.head()
```

### *Replacing the Square Brackets*

```python
data['ClientAge'] = data['ClientAge'].str.replace('[', '')
data['ClientDuration'] = data['ClientDuration'].str.replace(']', '')
```

```python
data.head()
```

### *Lowercasing the ItemDescription*

```python
data['ItemDescription'] = data['ItemDescription'].str.lower()
data.head()
```

## <center>**Joining in Python**
### *Importing new data*

```python
data2 = pd.read_csv('value_inc_seasons.csv', sep= ';')
data2.head()
```

### *Joining / Merging*

```python
merge_data = pd.merge(data, data2, on='Month')
merge_data.info()
```

## <Center>**Deleting variables from the dataframe**

```python
merge_data = merge_data.drop(['Year', 'Month', 'Day', 'ClientKeywords'], axis=1)
```

```python
merge_data.info()
```

## <center>**Sorting The Columns of the Merged Dataframe**

```python
merge_data = merge_data[['UserId', 'TransactionId', 'Country', 'Date', 'Season', 'Time', 'ClientAge', 'ClientType', 'ClientDuration', 'ItemCode', 'ItemDescription', 
'NumberOfItemsPurchased', 'CostPerItem', 'SellingPricePerItem', 'CostPerTransactions', 'SellingPricePerTransactions', 'ProfitPerItem', 'ProfitPerTransactions', 
'MarkupPerItem', 'MarkupPerTransactions']]
```

```python
merge_data.head()
```

## <center>**Exporting the dataframe**

```python
merge_data.to_csv('ValueIncCleaned.csv', index= False)
```
